### Задача №2
Если $P$ и $Q$ иррефлексивны, то $P \cup Q$, $P \cap Q$ и $P^{−1}$ таковы же.

Отношение $R$ множества $A$ иррефлексивно, если $id(A) \nsubseteq R$.

__1)__ Объединение множеств.<br>
Так как $P$ и $Q$ не содержат ни одного элемента из $id(A)$ (где $A$ — множество, на котором заданы отношения $P$ и $Q$), то их объединение так же не будет содержать таких элементов.

__2)__ Пересечение множеств.<br>
Так как $P$ и $Q$ не содержат ни одного элемента из $id(A)$, то их пересечение так же не будет содержать таких элементов.

__3)__ Обратное множество.<br>
Возведение множества в степень $-1$ воздействует только на существующие элементы, но не добавляет новых. Так как $P$ не подержит ни одного элемента из $id(A)$, то $P^{-1}$ тоже не содержит ни одного такого элемента.
