### Задача №2
Если $P$ и $Q$ иррефлексивны, то $P \cup Q$, $P \cap Q$ и $P^{-1}$ таковы же.

Отношение $R$ множества $A$ иррефлексивно, если $id(A) \nsubseteq R$.

__1)__ Объединение множеств.<br>
Так как $P$ и $Q$ не содержат ни одного элемента из $id(A)$ (где $A$ — множество, на котором заданы отношения $P$ и $Q$), то их объединение так же не будет содержать таких элементов.

__2)__ Пересечение множеств.<br>
Так как $P$ и $Q$ не содержат ни одного элемента из $id(A)$, то их пересечение так же не будет содержать таких элементов.

__3)__ Обратное множество.<br>
Возведение множества в степень $-1$ воздействует только на существующие элементы, но не добавляет новых. Так как $P$ не подержит ни одного элемента из $id(A)$, то $P^{-1}$ тоже не содержит ни одного такого элемента.

### Задача №4
Найдите в ч. у. м. $(P(\mathbb{N}), \subseteq)$ непустую цепь, где нет ни наибольшего, ни наименьшего элемента.

Возьмём подмножество $\mathbb{N}$ чётных чисел. Назовём его $\mathbb{N}_c$.

Тогда мы можем формировать наименьшие элементы до бесконечности, вычитая чётные числа __(*)__:
1. $\mathbb{N}_c - \{ 2 \}$;
1. $\mathbb{N}_c - \{ 2, 4 \}$;
1. $\mathbb{N}_c - \{ 2, 4, 6 \}$;
1. $\mathbb{N}_c - \{ 2, 4, 6, 8 \}$...

Аналогично с наибольшими элементами, только прибавляем нечётные __(#)__:
1. $\mathbb{N}_c + \{ 1 \}$;
1. $\mathbb{N}_c + \{ 1, 3 \}$;
1. $\mathbb{N}_c + \{ 1, 3, 5 \}$;
1. $\mathbb{N}_c + \{ 1, 3, 5, 7 \}$...

Цепь вида $\{ \textbf{(*)}, \mathbb{N}, \textbf{(#)} \}$ не имеет ни наибольшего, ни наименьшего элемента.

### Задача №6
Явно определите какой-либо _линейный_ порядок на множестве $\mathbb{R}^2$.

Множество $\mathbb{R}^2$ — это все пары вещественных чисел.

За линейный порядок можем взять операцию сравнения координат точки, где приоритет у первой координаты. Например:
1. (1, 1) < (1, 2);
1. (1, 1) < (2, 1);
1. (1, 2) < (2, 1);
1. (2, 2) > (2, 1)...

Таким образом, любые две точки на плоскости являются сравнимы, следовательно порядок линейный.
