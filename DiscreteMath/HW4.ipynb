За линейный порядок можем взять операцию сравнения координат точки, где приоритет у первой координаты. Например:
1. (1, 1) < (1, 2);

### Задача №8
Отношение $S$ на множестве $\mathbb{N}^2$ обладает следующим свойством: $(a, b)S(c, d) \Leftrightarrow (ad \neq bc$ и $b \neq 0 \neq d)$ или $(a = c$ и $b = 0 = d)$. Верно ли, что $S$ — отношение эквивалентности?

Чтобы проверить это утверждение, требуется проверить 3 свойства:
1. рефлексивность;
1. транзитивность;
1. симметричность.

__1.__<br>
$(a, b)S(c, d) \Leftrightarrow (ad \neq bc$ и $b \neq 0 \neq d)$ или $(a = c$ и $b = 0 = d)$ __(*)__

Рассмотрим два случая:
1. $b = 0$, тогда вторая скобка true, всё выражение тоже true;
2. $b \neq 0$, тогда вторая скобка true, всё выражение тоже true.

__2.__<br>
$(a, b)S(c, d) \Leftrightarrow (ad \neq bc$ и $b \neq 0 \neq d)$ или $(a = c$ и $b = 0 = d)$ __(*)__
$(c, d)S(a, b) \Leftrightarrow (cb \neq da$ и $b \neq 0 \neq d)$ или $(c = a$ и $b = 0 = d)$ __(#)__

По коммутативности знака равенства и умножения видим, что __(*)__ = __(#)__.

__3.__<br>
$(c, d)S(e, f) \Leftrightarrow (cf \neq de$ и $d \neq 0 \neq f)$ или $(c = e$ и $b = 0 = f)$ __(+)__

Рассмотрим два случая:
1. $0 \neq d \neq b \neq f$, тогда из равенств $ad = bc$ и $cf = de$ $\Rightarrow \frac{adt}{b} = de \Rightarrow \frac{af}{b} = e \Rightarrow af = be$ __(-)__;
1. $0 = d = b = f$, тогда из равенств $a = c$ и $c = e$ $\Rightarrow a = e$ __(%)__;

При других значениях $b, d, f$ не могут одновременно выполняться __(*)__ и __(+)__.

Таким образом, получаем выражение $(a, b)S(e, f) \Leftrightarrow (\mathbf{(af \neq be)}$ __(-)__ и $b \neq 0 \neq f)$ или $(\mathbf{(a = e)}$ __(%)__ и $b = 0 = f)$

Мы доказали рефлексивность, транзитивность и симметричность, поэтому это отношение эквивалентности.

### Задача №9
Пусть $R$ — бинарное отношение на множестве $A$. Когда $R$ является и частичным порядком, и отношением эквивалентности одновременно?

Чтобы $R$ являлся и частичным порядком, и отношением эквивалентности, требуется, чтобы выполнялись следующие условия:
1. рефлексивность;
1. транзитивность;
1. симметричность;
1. антисимметричность.

Проанализируем, когда выполняются все 4 свойства.

Учитывая симметричность и антисимметричность: $xRy \cap yRx \Rightarrow x = y$. Значит, если если какими-либо $x$ и $y$ существует связь, то это петля.

Учитывая рефлексивность, каждая вершина содержит петлю.

Таким образом, $R$ является и частичным порядком, и отношением эквивалентности, когда $R = id(A)$.
